```python
import os
import shutil
# Calculate difficult images:
difficult_img = 0
difficult_img_flag = True

path = '../input/dota/train/original/labels/remap_labels'
yolo_path = '../input/dota/train/compressed/all/yolo'
img_path = '../input/dota/train/compressed/all/images'
out_path = '../output/dota/val/split/'

for file in os.listdir(path):
    file_id = file.split('.')[0][1:]
    difficult_img_flag = None
    ship_only = True
    with open(os.path.join(path,file)) as f:
        df = f.read()
        for line in df.split('\n'):
            if line == '': break
            _,_, _,_, _,_, _,_, name, difficult = line.split(' ')
            difficult = int(difficult)

            if name == 'airport' or name == 'plane' or name == 'helicopter':
                difficult_img_flag = False
                ship_only = False
                break

            if name == 'plane' or name == 'oiltank':
                ship_only = False

            if difficult_img_flag is None and difficult:
                if name == 'plane' or name == 'oiltank' or name == 'warship':
                    difficult_img_flag = True
                    ship_only = False

        file_id = file.split('.')[0][1:]

        if ship_only == True:
            # Shipy
            if os.path.isfile(img_path+f'/P{file_id}.png'):
                shutil.copyfile(img_path+f'/P{file_id}.png', out_path+'imgsh'+f'/P{file_id}.png',follow_symlinks=True)
                shutil.copyfile(yolo_path+'/'+file, out_path+'labsh/'+file,follow_symlinks=True)

        elif difficult_img_flag is None or difficult_img_flag is False:
            # Normal
            if os.path.isfile(img_path+f'/P{file_id}.png'):
                shutil.copyfile(img_path+f'/P{file_id}.png', out_path+'img'+f'/P{file_id}.png',follow_symlinks=True)
                shutil.copyfile(yolo_path+'/'+file, out_path+'lab/'+file,follow_symlinks=True)


        elif difficult_img_flag is True:
            # Hard
            shutil.copyfile(img_path+f'/P{file_id}.png', out_path+'hard'+f'/P{file_id}.png',follow_symlinks=True)
            shutil.copyfile(yolo_path+'/'+file, out_path+'hardl/'+file,follow_symlinks=True)


print("Doone")

```
